```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
wave = pd.read_csv("wave.csv").drop(['Unnamed: 0'], axis=1)
wave
```

```python
wave[0:2000].plot(y=['o_signal_i', 'o_signal_q'])
```

```python
plt.magnitude_spectrum(wave['o_signal_i'], color ='green', Fs=100e6, scale='dB') 
plt.ylim(-70,100)
plt.xlim(0,0.5e7)
```

```python
plt.magnitude_spectrum(wave['o_signal_q'], color ='green', Fs=100e6, scale='dB') 
plt.ylim(-70,100)
plt.xlim(0,0.5e7)
```

```python
wave['o_signal_sum'] = wave['o_signal_i'] + wave['o_signal_q']
plt.magnitude_spectrum(wave['o_signal_sum'], color ='green', Fs=100e6, scale='dB') 
plt.ylim(-70,100)
plt.xlim(0,0.5e7)
```

```python
wave[0:2000].plot(y=['o_signal_sum'])
```
